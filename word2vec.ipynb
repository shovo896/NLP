```python
from gensim.models import Word2Vec
import gensim
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
nltk.download('punkt')
import warnings

warnings.filterwarnings(action='ignore')
```

```python
import zipfile
import re
from pathlib import Path

zip_path = Path("Gutenburg.zip")
if not zip_path.exists():
    raise FileNotFoundError(f"Missing file: {zip_path}")

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    file_name = zip_ref.namelist()[0]
    with zip_ref.open(file_name) as file:
        content = file.read().decode('utf-8', errors='ignore')
        cleaned_text = content.replace("\n", " " )
        cleaned_text = re.sub(r"\s+", " ", cleaned_text).strip()
        print("File Upload")
```

```python
data=[]
for sentence in sent_tokenize(cleaned_text):
    words = word_tokenize(sentence)
    words_lower = [word.lower() for word in words]
    data.append(words_lower)
```

```python
model1=gensim.models.Word2Vec(data,min_count=1,vector_size=100,window=5)
```

```python
model2=gensim.models.Word2Vec(data,min_count=1,vector_size=200,window=5,sg=1)
```

```python
print("cosine similarity between 'alice' and 'wonderland' - CBOW:", model1.wv.similarity('alice','wonderland'))
```

```python
print("cosine similarity between 'alice' and 'machine' - SG :", model2.wv.similarity('alice', 'machine'))
```
