```python
from transformers import T5Tokenizer,T5ForConditionalGeneration
import torch
```

```python
model_name = 't5-small'
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)
```

```python
input_text="translate English to German: The house is wonderful."
inputs=tokenizer.encode(input_text, return_tensors="pt")
```

geenrating output summary

```python
summary_ids=model.generate(inputs, max_length=40, num_beams=4, early_stopping=True)
output_text=tokenizer.decode(summary_ids[0], skip_special_tokens=True)
print(output_text)
```

performing translation 

```python
input_text="translate English to German: The house is wonderful."
inputs=tokenizer(input_text, return_tensors="pt")
translation_ids=model.generate(inputs.input_ids,max_length=40,num_beams=4,early_stopping=True)
translation_text=tokenizer.decode(translation_ids[0], skip_special_tokens=True)
print(translation_text)
```
