```python
from  nltk import word_tokenize,sent_tokenize
sent="Hello Mr. Smith, how are you doing today? The weather is great, and Python is awesome. The sky is pinkish-blue. You shouldn't eat cardboard."
print(sent_tokenize(sent))
print(word_tokenize(sent))
```

stemming 

```python
from nltk.stem import PorterStemmer
porter=PorterStemmer()
print(porter.stem("running"))
print(porter.stem("easily"))
```

```python
#example 
from nltk.stem import LancasterStemmer
lancaster=LancasterStemmer()
print(lancaster.stem("running"))
print(lancaster.stem("easily"))
```

```python
from nltk.stem import WordNetLemmatizer 
lemmatizer=WordNetLemmatizer()
print(lemmatizer.lemmatize("running","v"))
print(lemmatizer.lemmatize("easily","r"))
print(lemmatizer.lemmatize("better","a"))
```

```python
from nltk import pos_tag 
from nltk import word_tokenize
text="They refuse to permit us to obtain the refuse permit"
print(pos_tag(word_tokenize(text)))
```

```python
from nltk import pos_tag,word_tokenize,ne_chunk
nltk.download('maxent_ne_chunker')
nltk.download('words')
text="They refuse to permit us to obtain the refuse permit"
tokens=word_tokenize(text)
tagged=pos_tag(tokens)
print(ne_chunk(tagged))
```

```python
# Inspect spaCy pipeline
import spacy

# Try to load the small English model; fallback to blank if unavailable
try:
    nlp = spacy.load("en_core_web_sm")
except Exception:
    try:
        import sys, subprocess
        subprocess.check_call([sys.executable, "-m", "spacy", "download", "en_core_web_sm"])
        nlp = spacy.load("en_core_web_sm")
    except Exception:
        nlp = spacy.blank("en")

print(nlp.pipe_names)
```

```python
import spacy 
doc=nlp("Apple is looking at buying U.K. startup for $1 billion")
for token in doc:
    print(token,"|",token.idx,"|",token.pos_,"|",token.dep_,"|",token.shape_,"|",token.is_alpha,"|",token.is_stop,"|",token.lemma_)
```
