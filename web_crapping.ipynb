```python
import requests
from bs4 import BeautifulSoup

url="https://en.wikipedia.org/wiki/July_36"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

page_title=soup.title.text 
print("Page Title:", page_title)

# main heading 
main_heading=soup.find("h1").text
print("Main Heading:", main_heading)

# find paragraphs 
first_para=soup.find("p").text
print("First Paragraph:", first_para)

#all section headings 
print("\nSection Headings:")
for headings in soup.find_all(["h2","h3"]):
    print(headings.text.strip())
```
