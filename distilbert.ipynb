```python
from datasets import load_dataset
dataset = load_dataset("imdb")
train_dataset,test_dataset=dataset["train"],dataset["test"]
```

preprocessing  the data

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")

def process_function(examples):
    return tokenizer(examples["text"], truncation=True, padding="max_length", max_length=512)
tokenized_train=train_dataset.map(process_function, batched=True)
tokenized_test=test_dataset.map(process_function, batched=True)
```

load the pretrtained distilBert model

```python
from transformers import AutoModelForSequenceClassification

model=AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased", num_labels=2)
```

```python
from transformers import Trainer , TrainingArguments
training_args=TrainingArguments(output_dir="./results",eval_strategy="epoch",learning_rate=2e-5,per_device_train_batch_size=4,per_device_eval_batch_size=16,num_train_epochs=3,weight_decay=0.01,logging_dir='./logs',logging_steps=10,)
trainer=Trainer(model=model,args=training_args,train_dataset=tokenized_train,eval_dataset=tokenized_test,tokenizer=tokenizer,)
trainer.train()
trainer.evaluate()
```
