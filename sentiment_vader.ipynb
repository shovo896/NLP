```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
def  sentiment_scores(sentence):
    sid_obj= SentimentIntensityAnalyzer()
    sentiment_dict= sid_obj.polarity_scores(sentence)
    print(f"sentiment Scores:{sentiment_dict}")
    print(f"Negative score: {sentiment_dict['neg']*100}%")
    print(f"Neutral score: {sentiment_dict['neu']*100}%")
    print(f"Positive score: {sentiment_dict['pos']*100}%")
    if sentiment_dict['compound'] >= 0.05 :
        print("Overall Sentiment is Positive")
    elif sentiment_dict['compound'] <= - 0.05 :
        print("Overall Sentiment is Negative")
    else :
        print("Overall Sentiment is Neutral")

```
